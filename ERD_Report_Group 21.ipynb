## **Airline Reservation System Report**  

### **Project Overview**  
The Airline Reservation System is designed to manage and automate key airline operations, including flight scheduling, passenger reservations, payment processing, and flight status tracking. The system ensures efficient data handling, secure transactions, and real-time updates. The ERD outlines the structure of the database and the relationships between the tables, providing a clear view of how data is organized and connected.  

---

### **Objective**  
The objective of this system is to develop an integrated platform that enables:  
1. Easy booking and management of flights.  
2. Secure payment processing.  
3. Real-time updates on flight statuses.  
4. Efficient management of flight crew and passenger information.  
5. Tracking of luggage and handling special offers.  

---

### **Database Structure**  
The system consists of **10 interrelated tables** that cover all major aspects of airline operations:  

1. **Airlines**  
   - Stores airline information such as name, country, and headquarters.  
   - Each airline can have multiple flights.  

2. **Airports**  
   - Contains details of airports, including city, country, and a unique airport code.  
   - Each airport can have multiple departing and arriving flights.  

3. **Flights**  
   - Stores details about flights, including departure and arrival airports, airline, status, and aircraft type.  
   - Connected to the Airlines and Airports tables via foreign keys.  

4. **Passengers**  
   - Stores passenger details like name, contact information, passport number, and nationality.  
   - Each passenger can have multiple reservations.  

5. **Reservations**  
   - Records booking details, including flight, passenger, seat, and class.  
   - Linked to the Passengers and Flights tables.  

6. **Crew**  
   - Stores crew information such as name and role (Pilot, Co-Pilot, Attendant).  
   - Each crew member is linked to a specific flight.  

7. **Payments**  
   - Records payment details including amount, payment method, and payment date.  
   - Linked to the Reservations table.  

8. **Luggage**  
   - Stores details about luggage, including type and weight.  
   - Linked to the Reservations table.  

9. **Flight_Status**  
   - Monitors the real-time status of flights (Delayed, Departed, Landed, Cancelled).  
   - Linked to the Flights table.  

10. **Offers_Discounts**  
   - Tracks offers and discounts available to passengers.  
   - Linked to the Passengers table.  

---

### **Relationships and Cardinality**  
1. **One-to-Many Relationships:**  
- One airline → Many flights.  
- One airport → Many departing and arriving flights.  
- One passenger → Multiple reservations.  
- One reservation → Multiple pieces of luggage and one payment.  
- One flight → Multiple crew members and status updates.  

2. **Many-to-Many Relationships:**  
- None – All complex relationships have been simplified into one-to-many using junction tables.  

---

### **System Highlights**  
1. Data is normalized to reduce redundancy and improve consistency.  
2. Foreign key constraints ensure data integrity and prevent orphan records.  
3. Status fields use predefined ENUM values to maintain consistency.  
4. Indexed fields (like email and airport codes) improve search efficiency.  
5. Efficient handling of large volumes of data with optimized storage.
